```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dt = pd.read_csv("merged_datasets.csv")
```

```python
dt.head()
```

```python
dt["city"].unique()
```

```python
dt["city"].value_counts()
```

```python
dt['city'].value_counts().plot(kind='barh')
```

```python
np.where(dt["city"].value_counts() < 100)
```

```python
dt["City"] = dt["city"] 
```

```python
dt["City"].value_counts().dtypes
```

```python
dt["City"].value_counts().index
```

```python
city = dt.groupby("City")["City"].agg("count")
```

```python
city_less_100 = city[city<100]
```

```python
dt["City"] = dt["City"].apply(lambda i:"Other" if i in city_less_100 else i)
```

```python
dt["City"].value_counts()
```

```python
dt['City'].value_counts().plot(kind='barh')
```

```python
plt.figure(figsize = (20,10))
plt.pie(dt["City"].value_counts(), labels = dt["City"].unique(), autopct = "%2.1f%%")
plt.show()
```

```python
plt.figure(figsize = (20,10))
sns.countplot(x= dt["City"])
```
